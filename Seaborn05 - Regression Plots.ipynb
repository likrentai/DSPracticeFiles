# Regression Plots
* The **regplot** function plots data and a linear regression model fit.
* The **lmplot** function plots data and regression model fits across a FacetGrid (a multi-plot grid for plotting conditional relationships).
* The **residplot** function plots the residuals of a linear regression by regressing y on x (via a robust or polynomial regression) and then draw a scatterplot of the residuals.
* Due to the similarities between these three functions, only **lmplot** will be covered in this tutorial.
* Note: The concepts of regression are not covered here. Please refer to the Machine Learning section.
* Seaborn API documentation: https://seaborn.pydata.org/api.html
* Seaborn colour palette keywords and list: https://seaborn.pydata.org/tutorial/color_palettes.html
___

```python
# Library Import + Data Load
import seaborn as sns
%matplotlib inline

t = sns.load_dataset('tips')
t.head()
```

___
## lmplot()
* A combination of regplot() and FacetGrid. It is intended as a convenient interface to fit regression models across conditional subsets of a dataset.
* Split plots based on different features are possible by using hue and colouring.

```python
sns.lmplot(data=t, x='total_bill', y='tip')
```

```python
sns.lmplot(data=t, x='total_bill', y='tip', hue='sex')
```

```python
sns.lmplot(data=t, x='total_bill', y='tip', hue='sex', palette='magma')
```

### Markers Use (Optional)
* All the three types of regression have parameters with matplotlib capabilities.
* One of such parameters, scatter_kws (shown below), uses dictionary. The example, s:100 refers to the size of the scatter plot.

```python
sns.lmplot(x='total_bill', y='tip', data=t, hue='sex', palette='mako', markers=['o','x'], scatter_kws={'s':20})
```

### Grid Use
* Columns and rows are indicated as parameters.
* Instead of using hue as the separation of different features, grids can be used as variable separations through columns and rows.

```python
sns.lmplot(data=t, x='total_bill', y='tip', col='sex')
```

```python
sns.lmplot(data=t, x="total_bill", y="tip", row="sex", col="time")
```

```python
# Combining col and hue
sns.lmplot(data=t, x='total_bill', y='tip', col='day', hue='sex', palette='flare')
```

### Aspect & Sizes
Aspect ratio and size adjustment for Seaborn figures with the **aspect** and **height** parameters.

```python
# Building from the above example.
sns.lmplot(data=t, x='total_bill', y='tip', col='day', hue='sex', palette='flare', aspect=0.7, height=5)
```
