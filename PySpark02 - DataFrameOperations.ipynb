## Data Filtering
* A large part of working with Dataframes involves with quick filtering of data based on conditions.
* SQL commands and Spark dataframe methods can be used to filter data.

```python
# SQL filter command
df.filter("Close<400").show()
```

```python
# SQL filter command with .select()
df.filter("Close<400").select('Open').show()
```

```python
# SQL filter command with .select()
df.filter("Close<400").select(['Open','Close']).show()
```

* Spark dataframe format: df["column name"] + comparison operator

```python
# Dataframe filter command
df.filter(df["Close"]<200).show()
```

```python
# Common mistake with multiple conditions
# ValueError
df.filter(df["Close"]<200 and df['Open']>200).show()
```

```python
# Multiple conditions: Use additional parenthesis to separate the statements
df.filter((df["Close"]<200) & (df['Open']>200)).show()
```

```python
# OR operator
df.filter((df["Close"]<200) | (df['Open']>200)).show()
```

```python
# AND NOT operator
df.filter((df["Close"]<200) & ~(df['Open']<200)).show()
```

```python
# Obtain a specific row
df.filter(df["Low"] == 197.16).show()
```

```python
# Get results as Python list object for data manipulation
df.filter(df["Low"] == 197.16).collect()
```

```python
x = df.filter(df["Low"] == 197.16).collect()
```

```python
x[0]
```

```python
# The nested structure returns a nested row object
type(x[0])
```

```python
# Turn row into dictionaries
r = x[0]
r.asDict()
```

```python
# Get value from a specific key in dictionary
r.asDict()['Volume']
```

```python
# Iteration of values in a dictionary
for z in x[0]:
    print(z)
```
