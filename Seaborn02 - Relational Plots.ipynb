# Relational Plots
* The **relplot** function provides figure-level interface for drawing relational plots onto a FacetGrid.
* The **scatterplot** function draws a scatter plot with possibility of several semantic groupings.
* The **lineplot** function draws a line plot with possibility of several semantic groupings.
* Seaborn API documentation: https://seaborn.pydata.org/api.html
* Seaborn colour palette keywords and list: https://seaborn.pydata.org/tutorial/color_palettes.html
___

```python
# Library Import + Data Load
import seaborn as sns
%matplotlib inline

t = sns.load_dataset('tips')
t.head()
```

___
## relplot()
* Provides access to several different axes-level functions that show the relationship between two variables with semantic mappings of subsets.
* The parameter, kind='scatter' is the default. The other option: kind='line'

```python
sns.relplot(data=t, x='total_bill', y='tip', hue='day')
```

### Grid Use
* Instead of using hue as the separation of different features, grids can be used as variable separations through columns and rows.

```python
sns.relplot(data=t, x="total_bill", y="tip", hue="day", col="time")
```

```python
sns.relplot(data=t, x="total_bill", y="tip", hue="day", col="time", row="sex")
```

```python
# Using the parameter, kind='line' as the relational plot. (Not a suitable plot!)
sns.relplot(data=t, kind="line", x="total_bill", y="tip", hue="day")
```

```python
# A different dataset for demo
f = sns.load_dataset("flights")
f.head()
```

```python
sns.relplot(data=f, kind="line", x="year", y="passengers", hue="month", palette='dark')
```

___
## scatterplot()

```python
# Compare this with the first example
sns.scatterplot(data=t, x="total_bill", y="tip", hue="day")
```

```python
sns.scatterplot(data=t, x="total_bill", y="tip", hue="day", style="time")
```

___
## lineplot()

```python
# Based on the f = sns.load_dataset("flights")
sns.lineplot(data=f, x="year", y="passengers")
```

```python
sns.lineplot(data=f, x="year", y="passengers", hue='month')
```

```python
# The parameter, style groups variable
sns.lineplot(data=f, x="year", y="passengers", hue="month", style="month", markers=True)
```
