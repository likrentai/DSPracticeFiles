# Dates and Timestamps
* Handling date and time information in datasets.
___

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("datestimes").getOrCreate()
```

```python
# df = spark.read.format('csv').option('header','true').load('PySparkDataSets/applestock.csv')
df = spark.read.csv('PySparkDataSets/applestock.csv', header=True, inferSchema=True)
df.printSchema()
df.show()
```

```python
# Show the first row with the column names
df.head()
```

```python
# Show the specified column
df.select(['Date']).show()
```

```python
# Import relevant functions for date & timestamp data
from pyspark.sql.functions import format_number,dayofmonth,hour,dayofyear,month,year,weekofyear,date_format
```

```python
# Shows only the day data
df.select(dayofmonth(df['Date'])).show()
```

```python
# Shows only the hour data (if any)
df.select(hour(df['Date'])).show()
```

```python
df.select(dayofyear(df['Date'])).show()
```

```python
# Shows only the month data
df.select(month(df['Date'])).show()
```

```python
# Shows only the year data
df.select(year(df['Date'])).show()
```

```python
df.withColumn("Year",year(df['Date'])).show()
```

```python
# Obtain the average closing price based on years
# Step 1: Create an additional column to display only years
# Step 2: GroupBy years based on the mean for average closing price
a = df.withColumn("Year",year(df['Date']))
a.groupBy("Year").mean()[['Year','avg(Close)']].show()
```

```python
# Format to two decimal places + Rename column
b = a.groupBy("Year").mean()[['Year','avg(Close)']]
b.select('Year',format_number('avg(Close)',2).alias("Mean Close")).show()
```
