# Categorical Data Plots
* Examines the distribution of data in categorical columns.
* Seaborn API documentation: https://seaborn.pydata.org/api.html
* Seaborn colour palette keywords and list: https://seaborn.pydata.org/tutorial/color_palettes.html
* Types of seaborn plot for categorical data:
    * catplot (previously factorplot)
    * boxplot
    * violinplot
    * stripplot
    * swarmplot
    * barplot
    * countplot
___

## Libraries Import & Data Load

```python
# Library Import + Data Load
import seaborn as sns
%matplotlib inline

t = sns.load_dataset('tips')
t.head()
```

___
## barplot and countplot
* Enables aggregate data to be obtained from a categorical feature in the dataset.
* A **barplot** is a general plot to allow aggregation of the categorical data based on a function with finding the mean value as the default.

```python
sns.barplot(data=t, x='sex', y='total_bill')
```

```python
import numpy as np
```

* The estimator object can be set as a function to convert a vector to a scalar (affects grouping).
* Default = mean (i.e. estimator=np.mean)

```python
sns.barplot(data=t, estimator=np.std, x='sex', y='total_bill')
```

* In a **countplot**, the estimator is explicitly counting the number of occurrences. Hence, only the x value is pass in as an argument.

```python
sns.countplot(data=t, x='sex')
```

___
## boxplot and violinplot

* Both boxplots and violinplots are used to shown the distribution of categorical data.
* A box plot (or box-and-whisker plot) shows the distribution of quantitative data for comparisons between variables or across levels of a categorical variable. 
* The box shows the quartiles of the dataset while the whiskers extend to show the rest of the distribution.
* The outliers points are based on a method of a function of the inter-quartile range.

```python
# A boxplot is suitable for presentation to management (white collar but low skill level)
sns.boxplot(data=t, x="day", y="total_bill", palette='rainbow')
```

```python
# Show the whole dataframe with horizontal orientation, orient='h'
sns.boxplot(data=t, palette='rainbow', orient='h')
```

```python
sns.boxplot(data=t, x="day", y="total_bill", hue="smoker", palette="Paired")
```

* A violin plot has a similar role to a box and whisker plot.
* A violin plot shows the distribution of quantitative data across several levels of one (or more) categorical variables for comparison purposes.
* Differ from a box plot (where all of the plot components correspond to actual datapoints), a violin plot features a kernel density estimation of the underlying distribution.

```python
sns.violinplot(data=t, x="day", y="total_bill", palette='rainbow')
```

```python
sns.violinplot(data=t, x="day", y="total_bill", hue='sex', palette='Set1')
```

```python
sns.violinplot( data=t, x="day", y="total_bill", hue='sex', split=True, palette='Set1')
```

___
## stripplot and swarmplot
* A **stripplot** draws a scatterplot where one variable is categorical.
* A **swarmplot** serves as a good complement to a box or violin plot in cases where all observations along with some representation of the underlying distribution need to be shown.

```python
# sns.stripplot(data=t, x="day", y="total_bill", hue='time', palette='Set1')
sns.stripplot(data=t, x="day", y="total_bill")
```

```python
# The parameter, jitter add a small number of random noise datapoints for higher density of distribution
sns.stripplot(x="day", y="total_bill", data=t, hue='time', palette='Set1', jitter=True)
```

```python
sns.stripplot(x="day", y="total_bill", data=t, jitter=True, hue='sex', palette='Set1')
```

```python
sns.stripplot(x="day", y="total_bill", data=t, jitter=True, hue='sex', palette='Set1')
```

* A **swarmplot** shows non-overlapping points adjusted only along the categorical axis. Thus, giving a better representation of the values distribution. Note: Not recommended for large datasets.
* A combination of a stripplot and a violin plot.

```python
sns.swarmplot(data=t, x="day", y="total_bill")
```

```python
sns.swarmplot(x="day", y="total_bill", hue='sex', data=t, palette="Set1")
```

___
## Categorical Plots Combination

```python
sns.violinplot(x="tip", y="day", data=t, palette='rainbow')
sns.swarmplot(x="tip", y="day", data=t, color='black', size=4)
```

___
## catplot (previously factorplot)
* The most general form of a categorical plot. Accepts the parameter, **kind** to choose the plot type.
* Keywords including box, violin, strip, swarm, bar, count, etc.

```python
sns.catplot(data=t, x='day', y='total_bill', kind='bar')
```

```python
sns.catplot(data=t, x='day', y='total_bill', kind='violin')
```
