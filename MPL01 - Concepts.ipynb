## The Object Oriented API Method (Figure object):
* Figure objects can be represented. Calling of methods or functions can be applied to the object.
* Suitable for a canvas with multiple plots.

Example one (One set of figure in one canvas):

```python
# Step 1: Create empty canvas
fig = plt.figure()

# Step 2: Positioning & Size: Add set of axes with left, bottom, width, height with range 0 to 1
ax = fig.add_axes([0.1, 0.1, 0.9, 0.9])

# Step 3: Plot on the axes (.set_xxx methods); the datapoints are from arrays x and y above.
ax.plot(x, y, 'purple')
ax.set_xlabel('x axis Title')
ax.set_ylabel('y axis Title')
ax.set_title('Plot title')
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0, 0, 0.6, 0.6]) # main axes
ax2 = fig.add_axes([0.7, 0, 0.6, 0.6]) # inset axes
ax1.plot()
ax2.plot()
```

Note: This method provides more control such as the placement location of plot axes and the addition of more than one axis to the figure.

Example Two (Two sets of figures in one canvas):

```python
# Step 1: Creates empty canvas + Add axes (positioning & size; RANGE 0 - 1; try different numbers to see how they work!)
fig = plt.figure()

ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # main axes
ax2 = fig.add_axes([0.18, 0.52, 0.4, 0.3]) # inset axes

# Step 2: Set the axes for the big figure + datapoints from above
ax1.plot(x, y, 'brown')
ax1.set_xlabel('x axis big')
ax1.set_ylabel('y axis big')
ax1.set_title('Big plot title')

# Step 3: Set the axes for the small figure + datapoints from above
ax2.plot(y, x, 'orange')
ax2.set_xlabel('x axis small')
ax2.set_ylabel('y axis small')
ax2.set_title('Small plot title');
```

## .subplots()
* Multiple plots in a single plot display canvas.
* The method automatically manages the axis.

```python
# Not plt.figure() but plt.subplots() + use tuple unpacking to set fig and axes
fig, ax = plt.subplots()

# Use the axes object to add plot
ax.plot(x, y, 'red')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title');
```

* Specify the number of rows and columns when creating the subplots() objects

Note: Overlapping subplots or figures is a common problem with matplolib. To overcome the problem, the method **fig.tight_layout()** or **plt.tight_layout()** can be used to automatically adjusts the positions of the axes on the figure canvas so that there is no overlapping content.

```python
# To redo the above with indexing
fig, ax = plt.subplots(nrows=1, ncols=2)

ax[0].plot(x, y, 'gray')
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')
ax[0].set_title('Plot One')

ax[1].plot(y, x, 'seagreen')
ax[1].set_xlabel('x')
ax[1].set_ylabel('y')
ax[1].set_title('Plot Two')

plt.tight_layout()
```

## Figure size, aspect ratio and DPI

The aspect ratio, DPI and figure size can be specified when a figure object is created. 

The keyword arguments: `figsize` and `dpi`.
* `figsize` = a tuple of the width and height of the figure in inches
* `dpi` = the dots-per-inch (pixel per inch)

```python
fig = plt.figure(figsize=(8,4), dpi=100)
```

Example with layout managers including the `subplots` function:

```python
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(10,3))

ax[0].plot(x, y, 'cyan')
ax[1].plot(x, y, 'forestgreen')

plt.tight_layout()
```

## Customisation: Colours, Linewidths, and Linetypes
* Colours with MatLab-like syntax such as `'b'` for blue, `'g'` for green, etc. are accepted.
* CSS colour keywords such as 'seagreen', 'royalblue' are accepted.
* See the [documentation page](https://matplotlib.org/stable/gallery/color/named_colors.html) for all accepted colours.

```python
# Example: MATLAB style line color and style 
fig, ax = plt.subplots()
ax.plot(x, x**2, 'b.-') # blue line with dots
ax.plot(x, x**3, 'g--') # green dashed line
```

### Parameter, color='...'
* Colours can also be defined by RGB hex codes and optionally provide an alpha value using the `color` and `alpha` keyword arguments. Alpha indicates opacity.

```python
# Example with RGB hexcodes
fig, ax = plt.subplots()

ax.plot(x, x+1, color="blue", alpha=0.1) # 90% transparant
ax.plot(x, x+2, color="#8B008B")        # RGB hex code
ax.plot(x, x+3, color="#FF8C00")        # RGB hex code 
```

### Line and marker styles
* The `linewidth` or `lw` keyword argument changes the width of the plot line.
* The `linestyle` or `ls` keyword argument changes the style of the plot line.
* The use of markers is suitable when datapoints are relatively few.

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x+1, color="red", linewidth=0.25)
ax.plot(x, x+2, color="red", linewidth=0.50)
ax.plot(x, x+3, color="red", linewidth=1.00)
ax.plot(x, x+4, color="red", linewidth=2.00)

# possible linestype options -, , -., :, steps
ax.plot(x, x+5, color="seagreen", lw=3, linestyle='-')
ax.plot(x, x+6, color="seagreen", lw=3, ls='-.')
ax.plot(x, x+7, color="seagreen", lw=3, ls=':')

# custom dash
line, = ax.plot(x, x+8, color="black", lw=1.50)
line.set_dashes([5, 10, 15, 10]) # format: line length, space length, ...

# possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
ax.plot(x, x+ 9, color="royalblue", lw=3, ls='-', marker='+')
ax.plot(x, x+10, color="royalblue", lw=3, ls='--', marker='o')
ax.plot(x, x+11, color="royalblue", lw=3, ls='-', marker='s')
ax.plot(x, x+12, color="royalblue", lw=3, ls='--', marker='1')

# marker size and color
ax.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=2)
ax.plot(x, x+14, color="purple", lw=1, ls='-', marker='o', markersize=4)
ax.plot(x, x+15, color="purple", lw=1, ls='-', marker='o', markersize=12, markerfacecolor="skyblue")
ax.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', markersize=8, markerfacecolor="pink", markeredgewidth=3, markeredgecolor="green");
```

## Plot range (Axis appearance)
Axes range configuration:
* The `.set_ylim()` and `.set_xlim()` methods set the axes limits.
* The `.axis('tight')` method autoset "tightly fitted" axes ranges.

```python
fig, ax = plt.subplots(1, 3, figsize=(12, 4))

ax[0].plot(x, x**2, x, x**3)
ax[0].set_title("default axes ranges")

ax[1].plot(x, x**2, x, x**3)
ax[1].axis('tight')
ax[1].set_title("tight axes")

ax[2].plot(x, x**2, x, x**3)
ax[2].set_ylim([0, 60])
ax[2].set_xlim([2, 5])
ax[2].set_title("custom axes range");
```
