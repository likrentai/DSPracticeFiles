```python
# Library Import + Data Load
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
t = sns.load_dataset('tips')
t.head()
```

___
## set_style()
* Set the parameters that control the general style of the plots.
* The style parameters control properties like the color of the background and whether a grid is enabled by default. 

```python
# Basic countplot (pass in arguments into parameters for styles)
sns.countplot(data=t, x='sex')
```

```python
# Try style keywords: dark, whitegrid, 
sns.set_style('darkgrid')
sns.countplot(data=t, x='sex')
```

```python
sns.set_style('ticks')
sns.countplot(data=t, x='sex', palette='deep')
```

## Utility function: despine()
* Remove the top and right spines from plot(s).

```python
# top & right are true by default; bottom & left are false by default
sns.set_style('ticks')
sns.countplot(data=t, x='sex', palette='deep')
sns.despine(left=True)
```

___
## Aspect & Size
* **plt.figure(figsize=(width, height))** can be used to change the size of most seaborn plots.
* Used parameters: size and aspect to control the size and aspect ratio of most seaborn grid plots.

```python
# A non-grid plot
plt.figure(figsize=(12,4))
sns.countplot(data=t, x='sex')
```

```python
# A grid type plot
sns.lmplot(data=t, x='total_bill', y='tip', aspect=3)
```

___
## Themeing: set_context()
* The default parameters are overriden.
* Set the parameters to control the scaling of plot elements.
* Affected properties including the size of the labels, lines, and other elements of the plot (but not the overall style).

```python
# Example: settings for a poster
sns.set_context('poster', font_scale=1.25)
sns.countplot(data=t, x='sex', palette='mako')
```
