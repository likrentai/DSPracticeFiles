## Data Preprocessing

### Problems encountered:
* Many missing values (NaN) present in this dataset including those in the target column ['Malaria cases reported'].
* Removing the columns with missing values will also remove most of the existing values.
* Removing the rows with missing values will result in unequal number of rows in the X and y arrays. Thus, unable to data split.

### Approach:
* For the target column (y array), replace the missing values with the mean value. Alternative filling options include 0, mode or median.
* For the training X array data (X_train & X_test), imputate to replace the missing values with the mean value.

```python
# Import sklearn libraries
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.metrics import mean_absolute_error
from xgboost import XGBRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.impute import SimpleImputer
```

```python
# Identify the columns that require to be removed (due to too many NaN)
df.isnull().sum()
```

```python
# Remove the columns with more than 60 missing values
cc1 = [col for col in df.columns if df[col].isnull().sum()>60]
df = df.drop(columns=cc1)
df.isnull().sum()
```

```python
# Create a target column (the y array) and fill the missing values with the mean value
y = df['Malaria cases reported'].fillna(df['Malaria cases reported'].mean())
y.isnull().sum()
```

```python
# Separate the y array from numerical predictors (X array)
# Note: The missing values in the X array will be imputed in the next stage. X.isnull().sum()
a = df.drop('Malaria cases reported', axis=1)
X = a.select_dtypes(exclude=['object'])

# Data Split: Train-Test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
X_train.head()
```

```python
# Create two new columns to replace X_train and X_test. All the missing values have now been placed by mean values.
# Imputation
imputer = SimpleImputer(strategy='mean')
imX_train = pd.DataFrame(imputer.fit_transform(X_train))
imX_test = pd.DataFrame(imputer.transform(X_test))

# Imputation swap the imputed columns into the X_train & X_test
imX_train.columns = X_train.columns
imX_test.columns = X_test.columns

imX_train.isnull().sum()
```

___
## Linear Regression

```python
# Model Training
lm = LinearRegression()
lm.fit(imX_train, y_train)
```

```python
# Prediction & Evaluation
lrp = lm.predict(imX_test)
print('LR MAE:', metrics.mean_absolute_error(y_test, lrp))
print('LR MSE:', metrics.mean_squared_error(y_test, lrp))
print('LR RMSE:', np.sqrt(metrics.mean_squared_error(y_test, lrp)))
print('LR R2 Score:', lm.score(imX_test,y_test))
```

```python
# Optional scatterplot and residual histogram
plt.scatter(y_test, lrp)
sns.displot(y_test-lrp)
```

___
## Gradient Boosting Regressor

```python
# Model Training
gbr = GradientBoostingRegressor(random_state=0)
gbr.fit(imX_train, y_train)
```

```python
# Prediction & Evaluation
gbrp = gbr.predict(imX_test)
print('GBR MAE:', metrics.mean_absolute_error(y_test, gbrp))
print('GBR MSE:', metrics.mean_squared_error(y_test, gbrp))
print('GBR RMSE:', np.sqrt(metrics.mean_squared_error(y_test, gbrp)))
print('GBR R2 Score:', gbr.score(imX_test,y_test))
```

```python
# Optional scatterplot and residual histogram
plt.scatter(y_test, gbrp)
sns.displot(y_test-gbrp)
```
